```python
#3
import numpy as np

from array import *
A=[[2,3,1],[1,2,1],[2,1,1]]

# Compute  matrix of cofactors
cof_A = np.zeros_like(A)
for i in range(3):
    for j in range(3):
        minor = np.delete(np.delete(A, i, axis=0), j, axis=1)
        cof_A[i, j] = (-1) ** (i + j) * np.linalg.det(minor)

# cofactors transpose =adjoint matrix
adj_A = cof_A.T
print(adj_A)
det=np.linalg.det(A)
#inv(A)=adj(a)/det
invv=adj_A/det
print("Inverse of matrix",invv)
#inbuilt inverse function
inv=np.linalg.inv(A)
print("Inverse of matrix",inv)
    
```
